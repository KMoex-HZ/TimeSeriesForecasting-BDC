## 4. Time Series Decomposition (Medium)

Use `statsmodels.tsa.seasonal_decompose` to decompose the time series (e.g., 'value' column if available). 
Ensure the data is in datetime index.

**Expected Output:** Plot the trend, seasonal, and residual components.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

#membuat 'Order Date' sebagai indeks DataFrame dan diurutkan berdasarkan waktu
df_ts = df.set_index('Order Date').sort_index()

#mengagregasikan data ke frekuensi bulanan (sales per bulan)
monthly_sales = df_ts['Sales'].resample('MS').sum()

#dekomposisi Deret Waktu dengan model multiplikatif dengan periode 12 bulan
decomposition = seasonal_decompose(monthly_sales, model='multiplicative', period=12)

# 4 plot Komponen-komponen Dekomposisi Original Series, Trend, Seasonal, dan Residual
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.suptitle('Dekomposisi Deret Waktu Penjualan Bulanan', y=1.02) 
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()
```

Dekomposisi deret waktu dilakukan pada data penjualan bulanan (`Sales`) untuk memisahkan komponen tren, musiman, dan sisa. Ini membantu kita memahami pola dasar dalam penjualan.

Dari hasil plot dekomposisi di atas, kita dapat mengamati:

* **Original Series (Data Asli):** Menunjukkan fluktuasi penjualan bulanan yang sebenarnya terjadi selama periode 2014-2017.

* **Trend (Tren):** Terlihat jelas adanya **tren yang cenderung naik** (pertumbuhan) sepanjang periode data, terutama dari pertengahan 2015 hingga akhir data. Ini mengindikasikan bahwa bisnis secara keseluruhan sedang mengalami pertumbuhan dalam jangka panjang.

* **Seasonal (Musiman):** Terdapat **pola musiman yang kuat dan berulang setiap tahun**. Penjualan cenderung mengalami puncak di sekitar **pertengahan tahun (sekitar Juni-Juli)** dan juga di **akhir tahun (sekitar November-Desember)**, dengan periode penurunan di antaranya. Pola ini konsisten dari tahun ke tahun.

* **Residual (Sisa/Acak):** Komponen sisa menunjukkan fluktuasi yang relatif acak dan kecil di sekitar angka 1.0. Ini mengindikasikan bahwa sebagian besar variasi dalam data penjualan dapat dijelaskan dengan baik oleh komponen tren dan musiman.

```python
#mengakses komponen individualnya
trend_component = decomposition.trend
seasonal_component = decomposition.seasonal
residual_component = decomposition.resid

# plot komponen trend
print("Plot Komponen Trend")
plt.figure(figsize=(10, 5)) 
trend_component.plot()
plt.title('Komponen Tren Penjualan Bulanan') 
plt.xlabel('Tanggal')
plt.ylabel('Sales (Nilai Uang)')
plt.grid(True)
plt.show()

#plot komponen musiman
print("\nPlot Komponen Musiman (lebih detail per bulan)")

seasonal_cycle = seasonal_component.tail(12)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(10, 5)) 
plt.plot(months, seasonal_cycle) 
plt.title('Pola Musiman Penjualan Bulanan (per Siklus Tahun)') 
plt.xlabel('Bulan')
plt.ylabel('Faktor Musiman')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

#plot komponen residual
print("\nPlot Komponen Residual")
plt.figure(figsize=(10, 5)) 
residual_component.plot(marker='o', linestyle='')
plt.title('Komponen Sisa (Residual) Penjualan Bulanan') 
plt.xlabel('Tanggal') 
plt.ylabel('Faktor Sisa')
plt.grid(True)
plt.show()
```

Untuk pemahaman yang lebih mendalam, setiap komponen dekomposisi divisualisasikan secara terpisah:

#### Komponen Tren

Grafik Komponen Tren di atas lebih jelas menunjukkan jalur pertumbuhan penjualan dari waktu ke waktu. Terlihat bahwa setelah stabil di awal 2014, tren mulai menunjukkan kenaikan yang stabil dan signifikan mulai dari akhir 2015 hingga akhir 2017, mencapai puncaknya di atas 60.000 (nilai uang). Ini menegaskan adanya pertumbuhan bisnis yang positif.

#### Komponen Musiman

Grafik Komponen Musiman yang detail ini memperjelas pola berulang dalam satu tahun:
* **Puncak Penjualan:** Penjualan cenderung melonjak tinggi (faktor di atas 1.0, bahkan mendekati 1.8) di bulan **Maret**, dan mencapai puncak tertinggi di bulan **September, November, dan Desember**.
* **Lembah Penjualan:** Sebaliknya, penjualan cenderung menurun (faktor di bawah 1.0, bahkan di bawah 0.6) di bulan **Januari, Februari, April, Juni, dan Agustus**.

#### Komponen Sisa (Residual)

Grafik Komponen Sisa menunjukkan fluktuasi yang tidak dapat dijelaskan oleh tren maupun musiman. ## 6. Forecast with ARIMA (Hard)

Use ARIMA to build a forecasting model for the given time series.

**Steps:**
- Split data into train/test
- Fit ARIMA using `statsmodels`
- Forecast and plot comparison vs actual

**Expected:** Plot of actual vs forecast.

Train & Test

Pada bagian ini, model *Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors* (SARIMAX) digunakan untuk membuat prediksi penjualan bulanan. SARIMAX dipilih karena data penjualan menunjukkan adanya tren dan pola musiman yang jelas, seperti yang telah diidentifikasi pada tahap dekomposisi.

Data `monthly_sales` dibagi secara kronologis menjadi data latih (*train*) dan data uji (*test*). Pembagian ini penting untuk melatih model menggunakan data historis dan kemudian menguji akurasinya pada data yang belum pernah dilihat.

* **Data Latih (Train):** Digunakan untuk melatih model, mencakup periode **Januari 2014 hingga Desember 2016** (total 36 bulan).
* **Data Uji (Test):** Digunakan untuk menguji akurasi prediksi, mencakup periode **Januari 2017 hingga Desember 2017** (total 12 bulan). Pemilihan 12 bulan terakhir ini bertujuan untuk menguji kemampuan model dalam meramalkan satu siklus musiman penuh.

```python
#import library
#menggunakan model sarimax
from statsmodels.tsa.statespace.sarimax import SARIMAX 
from sklearn.metrics import mean_squared_error
import numpy as np

#train dan test
#data dari awal sampai Desember 2016
train_data = monthly_sales.loc[:'2016-12-01'] 
#data dari Januari 2017 sampai akhir
test_data = monthly_sales.loc['2017-01-01':]  

print(f"\nUkuran Data Latih (Train): {len(train_data)} bulan")
print(f"Ukuran Data Uji (Test): {len(test_data)} bulan")
print(f"Periode Data Latih: dari {train_data.index.min().strftime('%Y-%m')} sampai {train_data.index.max().strftime('%Y-%m')}")
print(f"Periode Data Uji: dari {test_data.index.min().strftime('%Y-%m')} sampai {test_data.index.max().strftime('%Y-%m')}")
```

Modeling

```python
#menentukan dan melatih model SARIMAX
arima_order = (1,1,1) 
arima_seasonal_order = (1,1,1,12)

print(f"\nMembuat model SARIMAX dengan order non-musiman {arima_order} dan order musiman {arima_seasonal_order}\n")

model = SARIMAX(train_data,
                order=arima_order,
                seasonal_order=arima_seasonal_order,
                enforce_stationarity=False, #False karena sudah pakai differencing (d dan D)
                enforce_invertibility=False)

results = model.fit(disp=False)
print(results.summary())
```

Model SARIMAX dilatih menggunakan data latih dengan parameter non-musiman `(1,1,1)` dan parameter musiman `(1,1,1,12)`. Parameter ini dipilih sebagai titik awal yang umum untuk data deret waktu dengan tren dan musiman bulanan/tahunan.

Selama proses pelatihan, muncul peringatan:
`UserWarning: Too few observations to estimate starting parameters for seasonal ARMA. All parameters except for variances will be set to zeros.`

Peringatan ini mengindikasikan bahwa jumlah data latih (36 bulan) mungkin belum cukup panjang bagi model SARIMAX untuk memperkirakan beberapa parameter musiman dan non-musiman awal secara optimal. Hal ini bisa mempengaruhi tingkat kepercayaan model terhadap pola-pola yang ditemukan.

Prediksi dan Plot

Setelah model dilatih, prediksi dilakukan untuk periode data uji (tahun 2017). Berikut adalah plot perbandingan antara data latih, data aktual (uji), dan hasil prediksi:

```python
#prediksi model dan plot
start_index = len(train_data) 
end_index = len(monthly_sales) - 1 

#prediksi
forecast = results.predict(start=start_index, end=end_index, dynamic=False)

forecast.index = test_data.index

print("\nPlot Perbandingan Data Aktual vs. Prediksi")
plt.figure(figsize=(14, 7))
# Plot Data Latih
plt.plot(train_data.index, train_data, label='Data Latih (Train)', color='blue')
# Plot Data Aktual (Test)
plt.plot(test_data.index, test_data, label='Data Aktual (Test)', color='orange')
# Plot Hasil Prediksi
plt.plot(forecast.index, forecast, label='prediksi SARIMAX', color='green', linestyle='--')
plt.title('prediksi Penjualan Bulanan (SARIMAX)')
plt.xlabel('Tanggal')
plt.ylabel('Sales (Nilai Uang)')
plt.legend()
plt.grid(True)
plt.show()
```

Dari plot di atas, terlihat bahwa:
* Model SARIMAX **berhasil menangkap tren pertumbuhan** pada data penjualan. Garis prediksi (hijau putus-putus) secara umum mengikuti arah tren data aktual (oranye).
* Model juga **berusaha menangkap pola musiman**, meskipun terdapat beberapa deviasi yang signifikan pada puncak atau lembah tertentu (misalnya, ramalan tidak setinggi puncak aktual di akhir tahun 2017).

Mengukur Akurasi

```python
#mengukur akurasi prediksi
#RMSE
rmse = np.sqrt(mean_squared_error(test_data, forecast))
print(f"\nRoot Mean Squared Error (RMSE) prediksi: {rmse:.2f}")

#MAPE
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / (y_true + np.finfo(float).eps))) * 100

mape_value = mape(test_data, forecast)
print(f"Mean Absolute Percentage Error (MAPE) prediksi: {mape_value:.2f}%")
```

Untuk mengukur seberapa akurat prediksi model, digunakan metrik *Root Mean Squared Error* (RMSE) dan *Mean Absolute Percentage Error* (MAPE).

* **Root Mean Squared Error (RMSE):** `16260.70`
    * RMSE menunjukkan rata-rata selisih antara nilai aktual dan prediksi dalam unit yang sama dengan data penjualan (nilai uang). Angka ini mengindikasikan bahwa rata-rata prediksi meleset sekitar $16.260 dari nilai aktual.

* **Mean Absolute Percentage Error (MAPE):** `26.84%`
    * MAPE menunjukkan rata-rata persentase selisih antara nilai aktual dan prediksi. Angka 26.84% berarti rata-rata prediksi meleset sekitar 26.84% dari nilai penjualan sebenarnya.

**Kesimpulan:**
Meskipun model SARIMAX berhasil menangkap tren dan sebagian besar pola musiman, akurasi prediksinya (dengan RMSE $16.260 dan MAPE 26.84%) masih menunjukkan adanya ruang untuk peningkatan. Hal ini mungkin disebabkan oleh keterbatasan panjang data latih, seperti yang ditunjukkan oleh *UserWarning* selama pelatihan model, atau parameter model yang belum sepenuhnya optimal.

## 7. Build Forecasting Model with Prophet (Hard)

Use Facebook Prophet to build a forecasting model.

**Steps:**
- Prepare data in Prophet format (`ds`, `y`)
- Fit the model and forecast the next 30 days
- Visualize the forecast using Prophetâ€™s plot

**Expected:** Forecasted time series and trend components.

Menyiapkan Data

Pada bagian ini, model prediksi Facebook Prophet digunakan untuk memprediksi penjualan harian. Prophet adalah pilihan yang tepat untuk data deret waktu yang memiliki tren dan pola musiman yang kompleks, serta mampu menangani data yang tidak teratur atau memiliki nilai hilang (meskipun data kita tidak ada nilai hilang).

Prophet memerlukan data dalam format spesifik dengan dua kolom: `ds` (untuk tanggal/waktu) dan `y` (untuk nilai yang akan diramal). Karena tugas kita adalah meramal 30 hari ke depan, data transaksi asli diubah menjadi data penjualan harian.
pre
**Agregasi Data Harian:** Data penjualan asli yang merupakan transaksi individu (sekitar 9994 entri) diagregasikan (dijumlahkan) menjadi total penjualan per hari. Ini menghasilkan 1458 hari unik dari tanggal transaksi pertama hingga terakhir, termasuk hari-hari tanpa penjualan (yang nilainya menjadi nol).

**Format Kolom:** Kolom Order Date diubah namanya menjadi `ds`, dan kolom Sales menjadi `y`.

```python
#import library
from prophet import Prophet
from sklearn.metrics import mean_squared_error

#menyiapiapkan data dalam format Prophet (ds, y)
#agregasi data penjualan ke frekuensi harian
df_daily_sales = df.set_index('Order Date')['Sales'].resample('D').sum().reset_index()

#mengubah nama kolom menjadi 'ds' dan 'y'
df_prophet = df_daily_sales.rename(columns={'Order Date': 'ds', 'Sales': 'y'})

#memastikan kolom 'ds' bertipe datetime
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'])

print("\n5 Baris Pertama Data dalam Format Prophet (ds, y):")
print(df_prophet.head())
print(f"\nUkuran Data Harian untuk Prophet: {len(df_prophet)} hari")
```

Train Test

Data harian dibagi secara kronologis untuk pelatihan dan pengujian model:

- **Data Latih (Train):** Digunakan untuk melatih model, mencakup periode 3 Januari 2014 hingga 31 Desember 2016 (total 1094 hari).
- **Data Uji (Test):** Digunakan untuk menguji akurasi prediksi, mencakup periode 1 Januari 2017 hingga 30 Desember 2017 (total 364 hari).


```python
train_prophet = df_prophet[df_prophet['ds'] <= '2016-12-31']
test_prophet = df_prophet[df_prophet['ds'] > '2016-12-31']

print(f"\nUkuran Data Latih (Train Prophet): {len(train_prophet)} hari")
print(f"Ukuran Data Uji (Test Prophet): {len(test_prophet)} hari")
print(f"Periode Data Latih Prophet: dari {train_prophet['ds'].min().strftime('%Y-%m-%d')} sampai {train_prophet['ds'].max().strftime('%Y-%m-%d')}")
print(f"Periode Data Uji Prophet: dari {test_prophet['ds'].min().strftime('%Y-%m-%d')} sampai {test_prophet['ds'].max().strftime('%Y-%m-%d')}")
```

Melatih Model Prophet

Model Prophet diinisialisasi dengan beberapa pengaturan penting:

- `seasonality_mode='multiplicative'`: Dipilih karena fluktuasi musiman dalam penjualan cenderung membesar seiring dengan peningkatan tren, sesuai dengan observasi dari dekomposisi.
- `daily_seasonality=False`: Tidak diaktifkan karena data harian penjualan cenderung tidak memiliki pola yang signifikan dalam skala jam-jaman.
- `weekly_seasonality=True`: Diaktifkan untuk menangkap pola penjualan yang berulang setiap minggu (misalnya, perbedaan penjualan antara hari kerja dan akhir pekan).
- `yearly_seasonality=True`: Diaktifkan untuk menangkap pola musiman tahunan yang telah teridentifikasi dari dekomposisi.

Model kemudian dilatih menggunakan data latih.

```python
#inisialisasi model
model_prophet = Prophet(seasonality_mode='multiplicative',
                        daily_seasonality=False,
                        weekly_seasonality=True,
                        yearly_seasonality=True)

#melatih model dengan data latih
model_prophet.fit(train_prophet)
```

Membuat DataFrame untuk Prediksi di Masa Depan dan Prediksi

Setelah model dilatih, Prophet digunakan untuk memprediksi penjualan. Prediksi dilakukan tidak hanya untuk periode data uji, tetapi juga untuk 30 hari ke depan setelah data historis terakhir.

Berikut adalah 5 baris terakhir dari hasil prediksi, menunjukkan tanggal (`ds`), nilai prediksi (`yhat`), serta rentang kepercayaan bawah (`yhat_lower`) dan atas (`yhat_upper`):

```python
#membuat DataFrame
future = model_prophet.make_future_dataframe(periods=30, freq='D', include_history=True)

#melakukan prediksi
forecast_prophet = model_prophet.predict(future)

print("\n5 Baris Pertama Hasil Prediksi (termasuk komponen):")
print(forecast_prophet[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()) # Menampilkan beberapa kolom penting
```

Visualisasi Prediksi

```python
#plot utama prediksi (data aktual vs. ramalan)
fig1 = model_prophet.plot(forecast_prophet)
plt.title('Prediksi Penjualan Harian dengan Prophet')
plt.xlabel('Tanggal')
plt.ylabel('Sales (Nilai Uang)')
plt.grid(True)
plt.show()
```

Dari plot utama, terlihat bahwa:

- Model Prophet berhasil menangkap tren pertumbuhan pada data penjualan harian. Garis prediksi (biru) secara umum mengikuti arah titik-titik data aktual (hitam).
- Prophet juga mampu menangkap fluktuasi harian/mingguan yang cepat pada data, ditunjukkan oleh garis prediksi yang naik-turun secara detail.
- Area biru muda menunjukkan rentang kepercayaan prediksi, yang memberikan gambaran tentang ketidakpastian model.

Visualisasi Komponen Prediksi Prophet

```python
#plot komponen-komponen Prediksi (tren, musiman mingguan, musiman tahunan)
fig2 = model_prophet.plot_components(forecast_prophet)
plt.suptitle('Komponen Prediksi Prophet', y=1.02) # Judul utama plot
plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Menyesuaikan layout
plt.show()
```

Prophet secara otomatis memisahkan dan memvisualisasikan komponen-komponen yang digunakannya untuk prediksi:

- **Trend:** Menunjukkan tren pertumbuhan jangka panjang yang stabil pada penjualan harian.
- **Weekly Seasonality:** Menggambarkan pola penjualan dalam satu minggu. Terlihat adanya puncak penjualan di hari Minggu dan Senin, serta di hari Jumat dan Sabtu, dengan penurunan signifikan di hari Rabu.
- **Yearly Seasonality:** Menggambarkan pola musiman penjualan dalam satu tahun. Pola ini mirip dengan hasil dekomposisi sebelumnya, menunjukkan puncak penjualan di sekitar Maret dan September-November, dengan penurunan di bulan-bulan lainnya.

Mengukur Akurasi Prediksi pada Data Uji

```python
#menggabungkan data aktual (test_prophet) dengan hasil prediksi (forecast_prophet) berdasarkan tanggal
forecast_test_period = forecast_prophet[forecast_prophet['ds'].isin(test_prophet['ds'])]
actual_test_period = test_prophet.set_index('ds').loc[forecast_test_period['ds']].reset_index()

#memastikan tidak ada NaN atau ketidaksesuaian indeks
if len(forecast_test_period) == len(actual_test_period):
    rmse_prophet = np.sqrt(mean_squared_error(actual_test_period['y'], forecast_test_period['yhat']))
    print(f"\nRoot Mean Squared Error (RMSE) Prophet pada data uji: {rmse_prophet:.2f}")

    mape_prophet = mape(actual_test_period['y'], forecast_test_period['yhat'])
    print(f"Mean Absolute Percentage Error (MAPE) Prophet pada data uji: {mape_prophet:.2f}%")
else:
    print("\nError: Jumlah data aktual dan prediksi pada periode uji tidak cocok. Tidak bisa menghitung metrik akurasi.")

#fungsi MAPE
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / (y_true + np.finfo(float).eps))) * 100
```

Akurasi model Prophet dievaluasi pada periode data uji (tahun 2017) menggunakan metrik Root Mean Squared Error (RMSE) dan Mean Absolute Percentage Error (MAPE).

**Root Mean Squared Error (RMSE): 1704.62**

RMSE ini menunjukkan rata-rata selisih antara nilai aktual dan prediksi dalam unit nilai penjualan ($) harian. Angka ini relatif rendah dibandingkan skala penjualan harian, mengindikasikan akurasi yang cukup baik.

**Mean Absolute Percentage Error (MAPE): nan%**

MAPE menghasilkan nan (Not a Number) karena adanya hari-hari dalam data uji yang memiliki nilai penjualan aktual nol. Rumus MAPE melibatkan pembagian dengan nilai aktual, yang akan menyebabkan error jika nilai aktualnya nol. Oleh karena itu, RMSE lebih relevan untuk evaluasi akurasi pada data harian dengan banyak nilai nol.

Model Prophet menunjukkan kinerja yang sangat baik dalam memprediksi penjualan harian, berhasil menangkap tren pertumbuhan, serta pola musiman mingguan dan tahunan yang kompleks. Meskipun ada peringatan MAPE, RMSE yang rendah mengkonfirmasi akurasi model yang kuat.
